## Section 1: Why Roles Matter

Models respond differently based on the identity you assign them. Role definition:

- **Focuses** the model on a specific domain or vertical
- **Sets expectations** for tone, expertise level, and response style
- **Provides context** that shapes all subsequent responses

Without a role, the model defaults to a generic "helpful assistant"â€”which may or may not be what you need.

````python
# Same question, no role vs with role
question = "What's wrong with this code? x = [1,2,3]; print(x[3])"

# No role
print("NO ROLE DEFINED:")
print("-" * 40)
response_no_role = call_mistral(user_prompt=question)
print(response_no_role)

print("\n" + "=" * 50 + "\n")

# With role
print("WITH ROLE DEFINED:")
print("-" * 40)
response_with_role = call_mistral(
    system_prompt="You are a senior Python developer who mentors junior developers. When explaining bugs, you explain the underlying concept, show the fix, and give tips to avoid similar issues.",
    user_prompt=question
)
print(response_with_role)
````

## Section 2: The Basic Pattern

Mistral recommends starting with this simple pattern:

```
You are a <role>, your task is to <task>.
```

This two-part structure is powerful because it:
1. **Establishes identity** (who the model is)
2. **Defines purpose** (what the model should do)

### Examples:

- "You are a **financial analyst**, your task is to **analyze quarterly reports and provide investment insights**."
- "You are a **language tutor**, your task is to **help students learn Spanish through conversation**."
- "You are a **technical writer**, your task is to **create clear documentation for APIs**."

````python
# Using the basic pattern
system_prompt = "You are a nutritionist, your task is to provide healthy meal suggestions based on dietary preferences."
user_prompt = "I'm vegetarian and trying to eat more protein. What should I have for dinner?"

response = call_mistral(system_prompt=system_prompt, user_prompt=user_prompt)

print(f"System: {system_prompt}\n")
print(f"User: {user_prompt}\n")
print(f"Response:\n{response}")
````

## Section 3: The Specificity Spectrum

More specific roles produce more focused outputs. Consider this progression:

| Level | Example | Quality |
|-------|---------|--------|
| Vague | "You are a helper" | Generic |
| Better | "You are a writing assistant" | Somewhat focused |
| Good | "You are a technical editor" | More focused |
| Best | "You are a technical editor specializing in API documentation for developer audiences" | Highly focused |

The more specific you are, the more the model can tailor its responses.

````python
# Let's see specificity in action
user_question = "How can I improve this paragraph?\n\nThe API lets you do stuff with data. You can get data and also put data. It's pretty fast."

specificity_levels = [
    ("Vague", "You are a helper."),
    ("Better", "You are a writing assistant."),
    ("Good", "You are a technical editor."),
    ("Best", "You are a technical editor specializing in API documentation for developer audiences. You focus on clarity, precision, and actionable details.")
]

for level, system_prompt in specificity_levels:
    print(f"Level: {level}")
    print(f"System: {system_prompt}")
    print("-" * 40)
    response = call_mistral(system_prompt=system_prompt, user_prompt=user_question)
    print(f"{response}")
    print("\n" + "=" * 50 + "\n")
````

## Section 4: Role Components

A well-defined role can include:

1. **Expertise area** - What they know
2. **Behavioral traits** - How they communicate
3. **Constraints** - What they won't do
4. **Audience awareness** - Who they're helping

### Example: Building a Complete Role

```
You are a senior data scientist with expertise in machine learning and statistics.
You communicate complex concepts in simple terms without being condescending.
You never make up statisticsâ€”if you don't know, you say so.
Your audience is business stakeholders who are not technical.
```

````python
# A fully specified role
system_prompt = """You are a senior data scientist with expertise in machine learning and statistics.
You communicate complex concepts in simple terms without being condescending.
You never make up statisticsâ€”if you don't know, you say so.
Your audience is business stakeholders who are not technical."""

user_prompt = "What's the difference between correlation and causation? Why does it matter for our business decisions?"

response = call_mistral(system_prompt=system_prompt, user_prompt=user_prompt)

print(f"Response:\n{response}")
````

---
## Section 5: Common Role Patterns

Here are patterns you can adapt for different use cases:

### Expert/Specialist
```
You are a senior [profession] with [X] years of experience in [specialty].
```

### Persona
```
You are a [adjective] [role] who [characteristic behavior].
```

### Critic/Reviewer
```
You are a [type] reviewer focused on [specific aspect]. Your feedback is [style].
```

### Transformer
```
You are a [role] that converts [input type] to [output type].
```

````python
# Different role patterns in action
user_input = "Our Q3 sales increased by 15% compared to Q2, mainly driven by the new product line launch in August."

role_patterns = {
    "Expert": "You are a senior business analyst with 15 years of experience in retail analytics.",
    "Persona": "You are an enthusiastic sales coach who celebrates wins while pushing for continuous improvement.",
    "Critic": "You are a skeptical analyst focused on finding gaps in business claims. Your feedback is direct and probing.",
    "Transformer": "You are a communications specialist that converts business data into social media posts."
}

for pattern_name, system_prompt in role_patterns.items():
    print(f"Pattern: {pattern_name}")
    print(f"Role: {system_prompt}")
    print("-" * 40)
    response = call_mistral(system_prompt=system_prompt, user_prompt=user_input)
    print(f"{response}")
    print("\n" + "=" * 50 + "\n")
````

---
## Section 6: What NOT to Do (Negative Examples)

### âŒ Too vague
```
You are helpful.
```
No specificityâ€”the model has no direction.

### âŒ Role doesn't match task
```
You are a medieval knight.
User: Help me debug this Python code.
```
Unless you want creative roleplay, match role to task.

### âŒ Contradictory traits
```
You are brief and concise. You always provide detailed, thorough explanations.
```
Pick one style.

### âŒ Overly complex role
```
You are a philosopher-scientist-artist-entrepreneur who thinks like Einstein, writes like Hemingway, and codes like Linus Torvalds...
```
Too many identities dilute focus.

````python
# Demonstrating a mismatched role
user_question = "What's the most efficient sorting algorithm for a nearly-sorted list?"

# Mismatched role
print("MISMATCHED ROLE:")
print("-" * 40)
response_mismatch = call_mistral(
    system_prompt="You are a medieval blacksmith who speaks in old English.",
    user_prompt=user_question
)
print(response_mismatch)

print("\n" + "=" * 50 + "\n")

# Matched role
print("MATCHED ROLE:")
print("-" * 40)
response_matched = call_mistral(
    system_prompt="You are a computer science professor specializing in algorithms.",
    user_prompt=user_question
)
print(response_matched)
````
